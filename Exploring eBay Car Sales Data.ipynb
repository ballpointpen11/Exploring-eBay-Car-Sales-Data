## Cleaning the Column Names

```python
# Creating a copy of the autos dataframe.
autos_copy = autos.copy()
autos_copy.columns

# Defining a function to change most of the column names.
def rename_columns(col):
    col = col.replace("yearOfRegistration","registration_year")
    col = col.replace("monthOfRegistration","registration_month")
    col = col.replace("notRepairedDamage","unrepaired_damage")
    col = col.replace("dateCreated","ad_created")
    col = col.replace("dateCrawled","date_crawled")
    col = col.replace("offerType","offer_type")
    col = col.replace("abtest","ab_test")
    col = col.replace("vehicleType","vehicle_type")
    col = col.replace("powerPS","power_PS")
    col = col.replace("fuelType","fuel_type")
    col = col.replace("nrOfPictures","nr_of_pictures")
    col = col.replace("postalCode","postal_code")
    col = col.replace("lastSeen","last_seen")
    return col

# Copying the 'autos.copy' columns names to the real 'autos' dataframe. 
renamed_columns = []
for c in autos.columns:
    renamed_col = rename_columns(c)
    renamed_columns.append(renamed_col)
autos.columns = renamed_columns
autos.head()
```

In the cell above, I have created a copy 'autos' to change the names of the columns.  These name changes will make the column labels more descriptive.  We have also changed the camelcased columns labels into snakecase.  Instead of each word starting with an upper case letter, all the labels are lowercase with underscores between each word.  After making the changes, I had to replace the column labels from 'autos' with the column labels from 'autos_copy'.

```python
# The describe method shows descriptive statistics for the columns that are 'int' dtypes.
autos.describe()
autos.head()
```

```python
print(autos['seller'].unique())
print(autos['nr_of_pictures'].unique())
print(autos['offer_type'].unique())
```

Above:  
- Three columns are candidates to get dropped for basically having a single value: 'seller', 'nr_of_pictures', and 'offer_type'.  'seller' only has 'privat' and 'nan' as values, 'nr_of_pictures' only has '0' and 'nan' as values, and 'offer_type' only has 'Angebot' and 'nan' as values.
- Two columns with numeric data stored as text are the 'price' and 'odometer' columns.  Those two columns need to have their non-numeric characters removed and need to be converted into numeric dtypes.  

## Dropping columns with mostly one value

```python
autos = autos.drop(['seller','nr_of_pictures','offer_type'], axis=1)
```

```python
autos.head()
```

We have now dropped the 'seller', 'nr_of_pictures', and 'offer_type' columns.  These were not helpful columns and will only clutter the dataset.

## 'Odometer' + 'Price' Converted into Numeric dtype

```python
# Renaming the 'odometer' column to 'odometer_km' to make it more descriptive.
autos.rename({"odometer":"odometer_km"},axis=1,inplace=True)

# Getting rid of the comma from both the 'price' and 'odometer_km' columns, and getting rid of the $ from the prices.  Also, switching the dtype of these two columns to float.
autos["price"] = autos["price"].str.replace("$","").str.replace(",","").astype(float)
autos["odometer_km"] = autos["odometer_km"].str.replace(",","").str.replace("km","").astype(float)
```

## Aggregating the brands

We will be aggregating the brands that have at least 5% of the total value in the dataset.

```python
brand_counts = autos["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > .05].index
print(common_brands)
```

There are 6 brands that have at least 5% of the total value in the dataset.

```python
aggregate_brands = {}

for b in common_brands:
    mean_price = autos.loc[autos['brand'] == b, 'price'].mean()
    aggregate_brands[b] = mean_price
```

```python
aggregate_brands
```

Volkswagen is the most represented car brand in the dataset. By mean price it is only more expensive than Opel and Ford cars.  The least expensive vehicle are Opel.  Audis are the third most represented car brand and the most expensive car by mean price.  

## Storing Aggregate Data in a Dataframe

For the top 6 brands, let's use aggregation to understand the average mileage for those cars and if there's any visible link with mean price.  First we need to calculate the mean mileage for each of the top brands, storing the results in a dictionary:

```python
mean_mileage = {}

for b in common_brands:
    mean_price = autos.loc[autos['brand'] == b, 'odometer_km'].mean()
    mean_mileage[b] = mean_price
mean_mileage
```

Now we have to convert both dictionaries (brands and mileage) to series objects, using the series constructor:

```python
# Converting 'aggregate_brands' into a series object.
brand_series = pd.Series(aggregate_brands)
brand_series = brand_series.sort_values(ascending=False)
brand_series
```

```python
# Converting 'mean_mileage' in a series object.
mileage_series = pd.Series(mean_mileage)
mileage_series = mileage_series.sort_values(ascending=False, inplace=False)
mileage_series
```

Now we have to create a datafram from the first series object using the dataframe constructor:

```python
brand_dataframe = pd.DataFrame(brand_series,columns=['mean_price'])
brand_dataframe = brand_dataframe
brand_dataframe
```

Now we have to assign the other series as a new column in the dataframe:

```python
# Adding a new column to 'brand_dataframe' filled with only zeroes for now.
brand_dataframe['mean_mileage_km'] = 0
brand_dataframe
```

```python
# Filling the rows of the column 'mean_mileage_km' with the data from 'mileage_series'.
brand_dataframe['mean_mileage_km'] = mileage_series
```

```python
brand_dataframe
```

The trend in this dataset shows that the cars with the higher mean prices tend to have a higher average mileage.  The top three most expensive cars: Audi, Mercedes Benz, and BMW have mileages of 127,487, 129,992, and 131,909, respectively.  The bottom three cars by mean price are Volkswagen, Ford, and Opel with mileages of 125,475, 119,099, and 123,588, respectively.  Overall the less expensive the car is, the less miles the car has.
